### scrapping data from makemytrip website

```python
HotelName = []
Address = []
Discount_Price = []
taxPrice = []
location = []
Rating =[]
Review=[]
Original_price=[]
cities = ['Mumbai','Goa','Ooty','Jaipur','Delhi','Manali','Bangalore','Shimla','Mahabaleshwar','Puri','alleppey','Hyderabad','Pune','Bangalore','chennai','coorg','lonavala','Gurgaon','Kolkata','Agra','Ajmer']
for city in cities:
    url ='https://www.makemytrip.com/hotels/{}-hotels.html'.format(city)
    page = requests.get(url,headers = {'User-Agent':'Mozilla/5.0'})
    page_context = page.text
    soup = BeautifulSoup(page_context)
    for x in soup.find_all('div', attrs={'class':'flexOne makeFlex'}):
        hname = x.find('span' , attrs={'class' : "wordBreak"  })
      
        rating = x.find('span' , attrs={'class' : "blackText latoBold"})
      
        address =  x.find('div' , attrs={'class' : "pc__html"})
      
        dis_p = x.find('p' , attrs={'class' : "latoBlack font26 blackText appendBottom5"})
        
        taxprice = x.find('p' , attrs={'class' : "font12 darkGreyText"})
        
        review= x.find('span' , attrs={'class' : "ratingText latoRegular font11"})
       
        org_p= x.find('p' , attrs={'class' : "grayText font16 lineThrough appendBottom5"})
    
        
        if hname is None:
            HotelName.append(np.NaN)
            location.append(city)
        else:
            HotelName.append(hname.text)
            
        if rating is None:
            Rating.append(np.NaN)
        else:
            Rating.append(rating.text)
        if address is None:
            Address.append(np.NaN)
        else:
            Address.append(address.text)

        if dis_p is None:
            Discount_Price.append(np.NaN)
        else:
            Discount_Price.append(dis_p.text)

        if taxprice is None:
            taxPrice.append(np.NaN)
        else:
            taxPrice.append(taxprice.text)
        
        if review is None:
            Review.append(np.NaN)
        else:
            Review.append(review.text)
          
        if org_p is None:
            Original_price.append(np.NaN)
        else:
            Original_price.append(org_p.text)
```

```python
print(len(HotelName),' ',len(Address),' ',len(Discount_Price),' ',len(Original_price),' ',len(taxPrice),' ',len(Rating),' ',len(Review))
```

```python
MMT = pd.DataFrame({
                    'HotelName' : HotelName , 
                    'Address':Address,                     
                    'Original_Rent':  Original_price,
                    'TaxPrice': taxPrice,
                    'Rating'  : Rating,
                    'Review' : Review,
                    'Discounted_Rent':Discount_Price,
    
                                  
                   })
```

## Bivariate plot for rent vs Quality

```python
fig = plt.figure(figsize=(12,9))
plt.ylabel('Rent', fontdict={'fontsize':20,'fontweight':'bold', 'color':'purple'})
plt.xlabel('Room Quality',fontdict={'fontsize':20,'fontweight':'bold', 'color':'purple'})
plt.title( 'Room Quality vs Rent',fontdict={'fontsize':25,'fontweight':'bold', 'color':'green'})
sns.boxplot(x='RoomQuality', y='Rent', data =cs, )
plt.show()
```


## Inference:
### From above graph it seems that rent is normally distributed as per room quality.
### There are some outliers, in room quality(Excellent) one of the hotels rent is approx 34000
### Some hotels of good quality  having rent in between 8000 to 12000.which is the budget friendly and very low
### hotels of very good  quality  having rent in between 12000 to 22000,and one of the hotel is having aprox 30000
